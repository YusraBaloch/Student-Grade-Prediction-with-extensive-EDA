```python
def common_comparison(params, grade='G1'):
    order_by = data.groupby(params)[grade].median().sort_values(ascending = False).index
    sns.boxplot(x = data[params], y = data[grade], order = order_by)
    plt.xticks(rotation=90)
    plt.title(f"{params} v/s {grade}")

parents_job = ['Fjob', 'Mjob']

plt.figure(figsize = (15 , 8))

for idx, name in enumerate(parents_job):
    plt.subplot(1, 2, idx+1)
    common_comparison(name, 'G3')

plt.show()
```

```python
parents_edu = ['Fedu', 'Medu']

plt.figure(figsize=(15, 8))

for idx, name in enumerate(parents_edu):
    plt.subplot(1, 2, idx+1)
    common_comparison(name, 'G3')
    plt.xticks(rotation=0)

plt.show()
```

```python
# Specifying IQ Range in order to Remove Outliers

num_columns = data.loc[:, data.columns != 'G3'].select_dtypes(include = np.number).columns.tolist()

Q1 = data[num_columns].quantile(0.25)
Q3 = data[num_columns].quantile(0.75)
IQR = Q3 - Q1

min = Q1 - 1.5 * IQR
max = Q3 + 1.5 * IQR

print('Quartile 1:\n', Q1)
print('\nQuartile 3:\n', Q3)
print('\nInter-Quartile Range:\n', IQR)
print('\nMin:\n', min)
print('\nMax:\n', max)
```

```python
# Remove Outliers

non_outlier_data = data.copy()
for i in num_columns:
  filter_min = non_outlier_data[i]<min[i]
  filter_max = non_outlier_data[i]>max[i]
  non_outlier_data = non_outlier_data[~(filter_min|filter_max)]

non_outlier_data.head()
```

```python
non_outlier_data.shape
```

```python
non_outlier_data.dtypes
```

```python
# ML Algorithms Implementation
```

```python
non_outlier_data['sex'] = non_outlier_data['sex'].astype('category')
non_outlier_data['sex_enc'] = non_outlier_data['sex'].cat.codes

non_outlier_data['address'] = non_outlier_data['address'].astype('category')
non_outlier_data['address_enc'] = non_outlier_data['address'].cat.codes

non_outlier_data['Mjob'] = non_outlier_data['Mjob'].astype('category')
non_outlier_data['Mjob_enc'] = non_outlier_data['Mjob'].cat.codes

non_outlier_data['Fjob'] = non_outlier_data['Fjob'].astype('category')
non_outlier_data['Fjob_enc'] = non_outlier_data['Fjob'].cat.codes

non_outlier_data['activities'] = non_outlier_data['activities'].astype('category')
non_outlier_data['activities_enc'] = non_outlier_data['activities'].cat.codes

non_outlier_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = non_outlier_data[['sex_enc', 'age', 'address_enc', 'Mjob_enc', 'Fjob_enc', 'studytime', 'failures', 'activities_enc', 'health', 'absences', 'G1', 'G2']]
y = non_outlier_data[['G3']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

lr = LinearRegression()
lr.fit(x_train, y_train)

predicted = lr.predict(x_test)

from sklearn.metrics import r2_score
ln_score = r2_score(y_test, predicted)

print('R2 Score Accuracy for Linear Regression: ', ln_score)
```

```python
from sklearn.ensemble import RandomForestRegressor

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
rf = RandomForestRegressor(n_estimators = 10, random_state = 0)
rf.fit (x_train, y_train)

rf_predicted = rf.predict(x_test)

rf_score = r2_score (y_test, rf_predicted)
print('R2 Score Accuracy for Random Forest Regressor: ', rf_score)
```

```python
import sklearn.neighbors as neighbors
class_model = neighbors.KNeighborsClassifier(n_neighbors = 10)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
class_model.fit(x_train, y_train)

KN_predict = class_model.predict(x_test)

KN_score = r2_score (y_test, KN_predict)
print('R2 Score Accuracy for KNeighbors Classifier: ', KN_score)
```

```python
from sklearn.linear_model import Ridge

estimator = Ridge(alpha=1, max_iter=10000)
estimator.fit(x_train, y_train)

ridge_predict = estimator.predict(x_test)

ridge_score = r2_score (y_test, ridge_predict)
print('R2 Score Accuracy for Ridge Classifier: ', ridge_score)
```

```python
from sklearn.tree import DecisionTreeRegressor

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)

dt_predict = dt.predict(x_test)

dt_score = r2_score(y_test , dt_predict)
print('R2 Score Accuracy for Decision Tree Regressor:', dt_score)
```

```python
ln = LinearRegression()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
ln.fit(x_train, y_train)

from sklearn.preprocessing import PolynomialFeatures
poly_model = PolynomialFeatures(degree=2)
x_poly = poly_model.fit_transform(x)

ln_2 = LinearRegression()
ln_2.fit(x_poly, y)

PR_pred = ln_2.predict(poly_model.fit_transform(x_test))

pr_score = r2_score(y_test, PR_pred)
print('R2 Score Accuracy for Polynomial Regressor:', pr_score)
```

```python
model_data = [['Polynomial Regression', pr_score], ['Ridge Classifier', ridge_score], ['Linear Regression', ln_score], ['Random Forest Regressor', rf_score], ['Decision Tree Regressor', dt_score], ['KNeighbor Classifier', KN_score]]
df_model = pd.DataFrame(model_data, columns = ['Algorithm', 'Score'])
df_model
```

```python
algorithm = list(df_model.Algorithm)
score = list(df_model.Score)

x = np.arange(len(algorithm))
width = 0.5

fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(15)
ax.set_xlabel("Algorithm")
ax.set_ylabel("Score")
ax.set_xticks(x)
ax.set_xticklabels(algorithm)

algo_frame = ax.bar(x, score, width, label='Algorithm')
for algo in algo_frame:
    height = algo.get_height()
    ax.annotate("{}".format(height),
                xy = (algo.get_x() + algo.get_width() / 2, height),
                xytext = (0,3),
                textcoords="offset points",
                ha='center', va='bottom')
    
plt.show()
```
